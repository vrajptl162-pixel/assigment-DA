```python
# 1.


import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/historical_automobile_sales.csv"
df = pd.read_csv(file_path)

# Display first few rows to understand structure
df.head()

# Assuming columns are 'Year' and 'Automobile_Sales'
# If different, adjust accordingly
print(df.columns)

# Group by Year and sum sales (if multiple entries per year)
yearly_sales = df.groupby("Year")["Automobile_Sales"].sum()

# Plot line chart
plt.figure()
plt.plot(yearly_sales.index, yearly_sales.values)
plt.xlabel("Year")
plt.ylabel("Automobile Sales")
plt.title("Year-wise Automobile Sales Trend")

plt.show()
```

```python
# 2.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('historical_automobile_sales.csv')

# Filter for recession periods
recession_data = df[df['Recession'] == 1]

# Group by Year and Vehicle_Type
df_line = recession_data.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Plotting the lines for different vehicle types
plt.figure(figsize=(8, 4))
sns.lineplot(data=df_line, x='Year', y='Automobile_Sales', hue='Vehicle_Type', marker='o')
plt.title('Automobile Sales Trends by Vehicle Type during Recession Periods')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.legend(title='Vehicle Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig('sales_trends_by_vehicle_type_recession.png')

# Display the head of the grouped data for verification
print(df_line.head())
```

```python
# 3.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv('historical_automobile_sales.csv')

# Group data by Recession and Vehicle_Type and calculate average Automobile_Sales
df_grouped = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the bar chart using Seaborn
# Map Recession values for better legend labels
df_grouped['Recession'] = df_grouped['Recession'].map({0: 'Non-Recession', 1: 'Recession'})

sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=df_grouped)

# Add titles and labels
plt.title('Average Automobile Sales by Vehicle Type: Recession vs Non-Recession')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.xticks(rotation=45)
plt.legend(title='Period')

# Save the plot
plt.savefig('sales_trend_comparison.png')
plt.close()

# Print the grouped data for verification in the response
print(df_grouped)
```

```python
# 4.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('historical_automobile_sales.csv')

# Inspect the data
print(df.head())
print(df.info())
print(df['Recession'].unique())
print(df['Vehicle_Type'].unique())

# Grouping the data
df_grouped = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Map Recession values for better labeling
df_grouped['Recession'] = df_grouped['Recession'].map({0: 'Non-Recession', 1: 'Recession'})

# Create the plot
plt.figure(figsize=(8, 4))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=df_grouped)
plt.title('Average Automobile Sales by Vehicle Type: Recession vs Non-Recession')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.xticks(rotation=45)
plt.legend(title='Period')
plt.tight_layout()
plt.savefig('sales_comparison_by_vehicle_type.png')

# Print the grouped data to see numeric differences
print(df_grouped)
```
