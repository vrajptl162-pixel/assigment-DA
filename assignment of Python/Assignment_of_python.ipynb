```python
# --- Q31: Count strings with same first and last character ---
def count_match(words):
    count = 0
    for word in words:
        if len(word) >= 2 and word[0] == word[-1]:
            count += 1
    return count
print("Matches found:", count_match(['abc', 'xyz', 'aba', '1221']))

# --- Q32: Remove duplicates from list ---
my_list = [10, 20, 30, 20, 10, 50]
unique_list = list(set(my_list))
print("List without duplicates:", unique_list)

# --- Q33: Check if list is empty ---
empty_list = []
if not empty_list:
    print("List is Empty")
else:
    print("List is not Empty")

# --- Q34: Common member in two lists ---
def common_data(list1, list2):
    result = False
    for x in list1:
        for y in list2:
            if x == y:
                result = True
                return result
    return result
print("Has common member:", common_data([1, 2, 3], [3, 4, 5]))

# --- Q35: Generate square of first 5 and last 5 numbers ---
def print_squares():
    l = list()
    for i in range(1, 31):
        l.append(i**2)
    print("First 5:", l[:5])
    print("Last 5:", l[-5:])
print_squares()

# --- Q36: Function taking list and returning unique elements ---
def unique_elements(l):
    x = []
    for a in l:
        if a not in x:
            x.append(a)
    return x
print("Unique:", unique_elements([1, 2, 3, 3, 3, 4, 5]))

# --- Q37: Convert list of characters to string ---
char_list = ['P', 'y', 't', 'h', 'o', 'n']
string_word = "".join(char_list)
print("String:", string_word)

# --- Q39: Create Tuple with different data types ---
my_tuple = ("Hello", 3.14, 100, True)
print("Tuple with mixed types:", my_tuple)

# --- Q40: Create a Tuple with numbers ---
num_tuple = (10, 20, 30, 40)
print("Number Tuple:", num_tuple)

# --- Q41: Convert tuple to string ---
tup = ('H', 'e', 'l', 'l', 'o')
str_tup = "".join(tup)
print("Tuple to String:", str_tup)

# --- Q42: Check if element exists in tuple ---
tup = ("apple", "banana", "cherry")
if "apple" in tup:
    print("Yes, apple is in the tuple")

# --- Q43: Length of a Tuple ---
print("Length of tuple:", len(tup))

# --- Q44: Convert list to tuple ---
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print("List to Tuple:", my_tuple)

# --- Q45: Reverse a tuple ---
tup = (1, 2, 3, 4)
rev_tup = tup[::-1]
print("Reversed Tuple:", rev_tup)

# --- Q46: Replace last value of tuples in a list ---
l = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
print("Original:", l)
# Lists are mutable, but tuples inside are not. We have to replace the whole tuple.
new_list = [t[:-1] + (100,) for t in l]
print("Modified List:", new_list)

# --- Q48: Sort Dictionary by value ---
import operator
d = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
# Ascending
sorted_d = dict(sorted(d.items(), key=operator.itemgetter(1)))
print("Sorted Dictionary (Ascending):", sorted_d)
# Descending
sorted_d_desc = dict(sorted(d.items(), key=operator.itemgetter(1), reverse=True))
print("Sorted Dictionary (Descending):", sorted_d_desc)

# --- Q49: Add key to dictionary ---
d = {0: 10, 1: 20}
d[2] = 30
print("Updated Dictionary:", d)

# --- Q50: Concatenate dictionaries ---
dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}
dic4 = {}
for d in (dic1, dic2, dic3):
    dic4.update(d)
print("Concatenated Dictionary:", dic4)

# --- Q51: Check if key exists in dictionary ---
d = {1: 10, 2: 20}
key = 2
if key in d:
    print(f"Key {key} exists")
else:
    print(f"Key {key} does not exist")

# --- Q53: Print dictionary where keys are 1-15 and values are squares ---
d = dict()
for x in range(1, 16):
    d[x] = x**2
print("Square Dictionary:", d)

# --- Q54: Merge two dictionaries ---
d1 = {'a': 100, 'b': 200}
d2 = {'x': 300, 'y': 200}
d = d1.copy()
d.update(d2)
print("Merged:", d)

# --- Q55: Map two lists into a dictionary ---
keys = ['red', 'green', 'blue']
values = ['#FF0000', '#008000', '#0000FF']
color_dict = dict(zip(keys, values))
print("Mapped Dictionary:", color_dict)
```

```python
# --- Q58/61: Combine values in list of dictionaries ---
from collections import Counter
item_list = [{'item': 'item1', 'amount': 400}, {'item': 'item2', 'amount': 300}, {'item': 'item1', 'amount': 750}]
result = Counter()
for d in item_list:
    result[d['item']] += d['amount']
print("Combined Dictionary:", result)

# --- Q60/62: Top 3 items in shop ---
from heapq import nlargest
my_dict = {'a': 500, 'b': 5874, 'c': 560, 'd': 400, 'e': 5874, 'f': 20}
three_largest = nlargest(3, my_dict, key=my_dict.get)
print("Top 3 keys:", three_largest)

# --- Q64: Function to calculate Factorial ---
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print("Factorial of 5:", factorial(5))

# --- Q65: Check if number is in given range ---
def test_range(n, start, end):
    if n in range(start, end+1):
        print(f"{n} is in the range {start}-{end}")
    else:
        print("The number is outside the given range.")
test_range(5, 1, 10)

# --- Q66: Check Perfect Number ---
def is_perfect(n):
    sum_n = 0
    for i in range(1, n):
        if n % i == 0:
            sum_n += i
    return sum_n == n
print("Is 6 perfect?", is_perfect(6))

# --- Q67: Check Palindrome (String) ---
def is_palindrome(s):
    return s == s[::-1]
print("Is 'madam' palindrome?", is_palindrome('madam'))

# --- Q69/70: Random Module Examples ---
import random

# Random item from list
color_list = ['Red', 'Blue', 'Green', 'White', 'Black']
print("Random Color:", random.choice(color_list))

# Random number from range
print("Random Number (1-100):", random.randint(1, 100))

# --- FILE HANDLING PREPARATION (Creating a dummy file for Colab) ---
# This code creates a file named 'test.txt' so the reading programs below work.
with open('test.txt', 'w') as f:
    f.write("Hello Python.\nThis is a test file.\nWe are learning file handling.")

# --- Q72: Read an entire text file ---
def file_read(fname):
    txt = open(fname)
    print("\n--- Entire File Content ---")
    print(txt.read())
file_read('test.txt')

# --- Q73: Append text to a file and display ---
def file_append(fname):
    from itertools import islice
    with open(fname, "a") as myfile:
        myfile.write("\nThis line is appended.")
    txt = open(fname)
    print("\n--- After Appending ---")
    print(txt.read())
file_append('test.txt')

# --- Q74: Read first n lines of a file ---
def file_read_from_head(fname, nlines):
    from itertools import islice
    with open(fname) as f:
        print("\n--- First 2 lines ---")
        for line in islice(f, nlines):
            print(line, end='')
file_read_from_head('test.txt', 2)

# --- Q76: Read a random line from a file ---
import random
def random_line(fname):
    lines = open(fname).read().splitlines()
    return random.choice(lines)
print("\n\nRandom Line:", random_line('test.txt'))

# --- Q80: Area of Shapes ---
import math

def area_trapezoid(base1, base2, height):
    return ((base1 + base2) / 2) * height

def area_parallelogram(base, height):
    return base * height

def area_cylinder(r, h):
    return (2 * math.pi * r * h) + (2 * math.pi * r**2)

print("Area of Trapezoid:", area_trapezoid(5, 6, 10))
print("Area of Parallelogram:", area_parallelogram(5, 10))
print("Surface Volume/Area of Cylinder:", area_cylinder(4, 5))

# --- Q85: Find Max and Min from Decimal numbers ---
def max_min_decimal(data):
    print("Maximum:", max(data))
    print("Minimum:", min(data))
max_min_decimal([1.1, 2.2, 5.5, 0.5])

# --- Q90: Exception Handling Example ---
try:
    x = 10 / 0 # This will cause an error
except ZeroDivisionError:
    print("\nError: Cannot divide by ZERO!")
finally:
    print("This 'finally' block always executes.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 1. Define the URL of the website
url = "http://books.toscrape.com/"

# 2. Send a request to fetch the data
response = requests.get(url)

# Check if the website is accessible (Status Code 200)
if response.status_code == 200:
    print("Connected to website successfully!")

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # List to store the data
    books_list = []

    # 3. Find all book items (inspected tag: article with class 'product_pod')
    all_books = soup.find_all('article', class_='product_pod')

    for book in all_books:
        # Extract Title
        title = book.h3.a['title']

        # Extract Price
        price = book.find('p', class_='price_color').text

        # Extract Rating (It's in the class name, e.g., "star-rating Three")
        rating = book.p['class'][1]

        # Add to list
        books_list.append({
            'Book Title': title,
            'Price': price,
            'Rating': rating
        })

    # 4. Convert to DataFrame and Save to CSV
    df = pd.DataFrame(books_list)

    # Display first few rows
    print("\n--- Extracted Data Preview ---")
    print(df.head())

    # Save file
    df.to_csv('scraped_books.csv', index=False)
    print("\nData saved to 'scraped_books.csv' successfully.")

else:
    print("Failed to connect. Please check the URL.")
```
